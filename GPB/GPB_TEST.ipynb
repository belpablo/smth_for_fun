# Анализ сбережений в банке N:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

xl = pd.ExcelFile('GPB LevelUp ЦО Сбережения.xlsx')
df = xl.parse(xl.sheet_names[0])
df
```

```python

names_arr = list(df.columns)
df = df[[names_arr[i] for i in [2, 1, 3, 4, 5, 6]]]

names_arr = list(df.columns)
df
```

```python
da_word = df[names_arr[5]][0]

for i in range(len(df)):
    
    df[names_arr[2]][i] = int(df[names_arr[2]][i] == da_word)
    df[names_arr[5]][i] = int(df[names_arr[5]][i] == da_word)
    
df
```

```python
for i in range(len(df)):
    x = int(df[names_arr[3]][i].split()[0])
    df[names_arr[3]][i] = int(x + 11 * (1 + np.sign(2 - x)) / 2) 
    
df
```

```python
# вот такая вот интересная функция, чтобы не писать лишний if :)
x = np.linspace(0, 10)
plt.plot(x,  11 * (1 + np.sign(2 - x)) / 2)
```

```python
pd.plotting.scatter_matrix(df,  diagonal = 'kde', figsize = (10, 10))
plt.show()
```

```python
import seaborn as sns

data = df[[names_arr[i] for i in [0, 1, 4]]]

fig = plt.figure(figsize = (10, 10))
g = sns.PairGrid(data[data.columns], vars = data.columns)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill = True)
g.map_diag(sns.histplot, kde = True)
```

```python
sample_arr = [1, 2]
title_arr = [names_arr[i] for i in sample_arr]

for i in sample_arr:

    data = df[df.columns[i]]
    arr = [0 for j in range(len(data.unique()))] 
    for num in data:
        arr[num - 1] += 1

    if i != 2:
        labels = data.unique()
    else:
        labels = ['да', 'нет']

    fig1, ax1 = plt.subplots()
    ax1.pie(arr, autopct = '%1.1f%%')
    patches, texts, auto = ax1.pie(arr, autopct = '%1.1f%%' )

    ax1.axis('equal')
    plt.legend(patches, labels, loc = "best")
    plt.title(title_arr[i - 1])
    plt.show()
```

# Анализ данных о банках для выбора оптимальной стратегии:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

xl = pd.ExcelFile('GPB LevelUp ЦО Общая вводная.xlsx')
df = xl.parse(xl.sheet_names[0])
df.head()
```

```python
df_cost = df[df.index % 2 == 0].reset_index(drop = True).drop(19, axis = 0)
df_cost = df_cost.rename(columns = {df_cost.columns[0]: 'Банк'})
df_cost = df_cost.sort_values('Банк').reset_index(drop = True)
names_arr = df_cost.columns

for i in range(len(df_cost)):
    df_cost[names_arr[0]][i] = df_cost[names_arr[0]][i].split()[1]
    
df_cost
```

```python
import matplotlib.colors as mcolors

#bc = mcolors.BASE_COLORS
#num_set = np.random.randint(1, len(mcolors.BASE_COLORS), len(df_cost))
#colors = [list(bc.keys())[i] for i in num_set]
fig = plt.figure(figsize = (12, 8))
#plt.yscale('log')

for i in range(len(df_cost)):
    plt.plot(names_arr[1:-1], df_cost.loc[i][1:-1], label = 'Банк ' + df_cost.loc[i][:1][0]) #, c = colors[i])
    #plt.plot(names_arr[1:-1], df_cost.loc[i][1:-1], '.', c = colors[i])

plt.legend(loc = 'best')
plt.grid()
```

Из приведённого выше графика видно, что из общей массы банков, только банк M обладает рыночной капитализацией и темпами роста с схожими с нашим банком (N) => логично учитывать их опыт изменения стратегии при выборе своей.

```python
df_grow = df[(df.index % 2 == 1)].reset_index(drop = True)
df_grow = df_grow.drop([names_arr[i] for i in [1, 13]], axis = 1).drop(19, axis = 0)
df_grow = df_grow.rename(columns = {df_grow.columns[0]: 'Банк'})
df_grow = df_grow.sort_values('Банк').reset_index(drop = True)
names_arr = df_grow.columns

for i in range(len(df_grow)):
    df_grow[names_arr[0]][i] = df_grow[names_arr[0]][i].split()[2]
df_grow
```

```python
import matplotlib.colors as mcolors

bc = mcolors.BASE_COLORS
sample_arr = [0, 1, 4, 5, 6, 7, 8]
colors = [list(bc.keys())[i - 8] for i in range(len(sample_arr))]

boo_years_arr = [2020, 2017, 2012, 2017, None, 2018, 2017]


fig = plt.figure(figsize = (12, 8))

for num, i in enumerate(sample_arr):
    plt.plot(names_arr[1:], df_grow.loc[i][1:], label = 'Банк ' + df_grow.loc[i][:1][0], c = colors[num])
    plt.plot(names_arr[1:], df_grow.loc[i][1:], '.', c = colors[num]) 
    if i == 0:
        plt.scatter(boo_years_arr[num] - 1, df_grow[boo_years_arr[num] - 1][i], c = 'k', marker = 'D', linewidths = 4, 
                    label = 'Bank-Of-One') 
    elif i != 6:
        plt.scatter(boo_years_arr[num] - 1, df_grow[boo_years_arr[num] - 1][i], c = 'k', marker = 'D', linewidths = 4) 


plt.title('Темпы роста банков (%)')
plt.legend(loc = 'best')
plt.grid()
```

Как видно из графика, введение системы Bank-of-one вызвало резкий скачок в темпах роста, однако данный подход оказался мало эффективным в епрспективе.

Также поступила информация, что банк M после введения сисетым bank-of-one просел по фин покаателям на 20 процентов (банк схожий с нашим по стратегии и капитализации) => эта система нам точно не подходит
