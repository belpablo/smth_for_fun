# Анализ сбережений в банке N:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

xl = pd.ExcelFile('GPB LevelUp ЦО Сбережения.xlsx')
df = xl.parse(xl.sheet_names[0])
df
```

```python

names_arr = list(df.columns)
df = df[[names_arr[i] for i in [2, 1, 3, 4, 5, 6]]]

names_arr = list(df.columns)
df
```

```python
da_word = df[names_arr[5]][0]

for i in range(len(df)):
    
    df[names_arr[2]][i] = int(df[names_arr[2]][i] == da_word)
    df[names_arr[5]][i] = int(df[names_arr[5]][i] == da_word)
    
df
```

```python
for i in range(len(df)):
    x = int(df[names_arr[3]][i].split()[0])
    df[names_arr[3]][i] = int(x + 11 * (1 + np.sign(2 - x)) / 2) 
    
df
```

```python
# вот такая вот интересная функция, чтобы не писать лишний if :)
x = np.linspace(0, 10)
plt.plot(x,  11 * (1 + np.sign(2 - x)) / 2)
```

```python
pd.plotting.scatter_matrix(df,  diagonal = 'kde', figsize = (10, 10))
plt.show()
```

```python
import seaborn as sns

data = df[[names_arr[i] for i in [0, 1, 4]]]

fig = plt.figure(figsize = (10, 10))
g = sns.PairGrid(data[data.columns], vars = data.columns)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill = True)
g.map_diag(sns.histplot, kde = True)
```

```python
sample_arr = [1, 2]
title_arr = [names_arr[i] for i in sample_arr]

for i in sample_arr:

    data = df[df.columns[i]]
    arr = [0 for j in range(len(data.unique()))] 
    for num in data:
        arr[num - 1] += 1

    if i != 2:
        labels = data.unique()
    else:
        labels = ['да', 'нет']

    fig1, ax1 = plt.subplots()
    ax1.pie(arr, autopct = '%1.1f%%')
    patches, texts, auto = ax1.pie(arr, autopct = '%1.1f%%' )

    ax1.axis('equal')
    plt.legend(patches, labels, loc = "best")
    plt.title(title_arr[i - 1])
    plt.show()
```
