### Write a program, that computes n first even Fibonacci numbers for each given n. For example:

$f(4) = [0, 2, 8, 34]$

### To complite the task let's start from some basics, like writing Fibbonaci function at it pure form: 

```python
def fib_num(i, nfev = 0):
    
    '''рекурентная функция, возвращающая i-е число Фибоначчи и число вызовов этой функции'''
    
    if i == 0:
        return [0, nfev + 1]
    elif i == 1:
        return [1, nfev + 1]
    else:
        first_prev = fib_num(i - 1)
        second_prev = fib_num(i - 2)
        return [first_prev[0] + second_prev[0], first_prev[1] + second_prev[1]]
    
for i in range(21):
    curr_arr = fib_num(i)
    print(str(i) + '-th Fibbonaci num: ' + str(curr_arr[0]) + ', nfev = ' + str(curr_arr[1]))
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt


def fib_num(i):
    
    '''рекурентная функция, возвращающая i-е число Фибоначчи'''
    
    if i == 0:
        return 0
    elif i == 1:
        return 1
    else:
        return fib_num(i - 1) + fib_num(i - 2)
    
start_num, stop_num = 15, 40
iter_arr, times_arr = [i for i in range(start_num, stop_num)], []

for i in iter_arr:
    curr_times_arr = []
    for j in range(5):
        strat_time = time.time()
        fib_num(i)
        end_time = time.time()
        curr_times_arr.append(abs(end_time - strat_time))
    times_arr.append(np.mean(curr_times_arr))

fig = plt.figure(figsize  = (12, 6))
plt.plot(iter_arr, times_arr, c = 'navy')
plt.xlabel('number of iterations', fontsize = 14)
plt.ylabel('time, sec', fontsize = 14)
plt.grid()
```

```python
fig = plt.figure(figsize  = (12, 6))
#plt.xlim(start_num, stop_num)
plt.yscale('log')
plt.plot([start_num + i for i in range(len(times_arr))], times_arr, c = 'navy')
plt.xlabel('number of iterations', fontsize = 14)
plt.ylabel('time, sec', fontsize = 14)
plt.grid()
```
